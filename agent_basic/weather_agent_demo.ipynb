## Configure the Agent

Set up the agent configuration and load the environment variables required for the OpenWeatherMap API key.


```python
from dotenv import load_dotenv
from weather_agent import WeatherAgent
import os

load_dotenv()
api_key = os.getenv("OPENWEATHERMAP_API_KEY")
if not api_key:
    raise RuntimeError(
        "OPENWEATHERMAP_API_KEY is not set. Copy .env.example to .env and add your key."
    )

agent = WeatherAgent(api_key)

# Example prompt
prompt = "What is the weather in Paris?"
print("Prompt:", prompt)
print(agent.handle_user_request(prompt))
```

## Run the Agent in the Notebook

This section shows how to execute the agent directly from notebook code. The agent is initialized and then used to process a prompt.


```python
# Run a few example interactions with the notebook agent
from dotenv import load_dotenv
from weather_agent import WeatherAgent
import os

load_dotenv()

api_key = os.getenv("OPENWEATHERMAP_API_KEY")
if not api_key:
    raise RuntimeError(
        "OPENWEATHERMAP_API_KEY is not set. Copy .env.example to .env and add your key."
    )

agent = WeatherAgent(api_key)

examples = [
    "What is the weather in London?",
    "Check weather for Tokyo.",
    "Current weather in New York",
]

for prompt in examples:
    print(f"Prompt: {prompt}")
    print(agent.handle_user_request(prompt))
    print()
```
