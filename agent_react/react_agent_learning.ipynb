## Understanding the ReAct pattern

ReAct stands for "Reasoning and Acting." It is a design where the agent alternates between:

- thinking about the problem,
- deciding whether to call a tool,
- calling that tool, and
- using the result to continue reasoning.

This keeps the reasoning process explicit and lets the agent use tools only when needed.

## What is an agent harness?

The agent harness is the core controller. It:

- builds the prompt sent to the model,
- decides which skill should handle the request,
- manages tools and tool calls,
- updates memory, and
- returns the final answer.

Think of it as the "brain" that guides the agent's behavior.

## What are AI skills?

AI skills are reusable capabilities that the agent can choose before it generates a response.

Each skill has:

- a name,
- a description,
- keywords that match user intent, and
- an optional corresponding tool.

This makes the agent more modular and easier to extend.

## MCP-style context separation

MCP stands for Message Channel Program. In this notebook, the agent keeps different kinds of information separate so the model can use them correctly.

These channels are:

- `system`: instructions about agent behavior,
- `memory`: recent conversation history and tool results,
- `tool descriptions`: what tools are available, and
- `AI skills`: which skill the agent has selected.

Keeping channels separate is a core harness design principle.

## Memory management

A simple agent keeps a short history of recent exchanges and tool outputs.

In this notebook, memory is used to help the agent remember:

- what the user asked,
- what the agent answered,
- and what each tool returned.

This keeps the conversation coherent across turns.

## Tool execution and observations

Tools are separate functions that the agent can call when it needs extra information or a calculation.

This notebook shows two tools:

- `search_knowledge` for looking up facts,
- `calculate` for numeric evaluation.

The harness also feeds tool outputs back into the reasoning loop so the agent can use them in final answers.

### How the agent chooses a skill

The harness examines the user query and selects one of the registered skills. In this example, the agent can choose between:

- `knowledge` — for facts, definitions, and concept lookups

- `general` — for open-ended questions and conversational replies

```python
query = 'What is the ReAct pattern and how does it differ from a simple prompt-only answer?'
answer = agent.run(query)
print('Selected skill:', agent.last_selected_skill.name)
print('Answer:')
print(answer)
```
